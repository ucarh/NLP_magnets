```python
def get_prediction(model_name, dataset): 
    '''model names could be only:
       - bert_squad, magbert_squad, magmatbert_squad, matscibert_squad
       - bert_squad_Tc, magbert_squad_Tc, magmatbert_squad_Tc, matscibert_squad_Tc'''
    if model_name == "bert_squad":
        model_checkpoint = "nlp-magnets/QA_best_bert-base-uncased_4_epochs_lr_3e-05_BS_16"
    elif model_name == "magbert_squad":
        model_checkpoint = "nlp-magnets/QA_best_magbert_4_epochs_lr_5e-05_BS_64"
    elif model_name == "magmatbert_squad":
        model_checkpoint = "nlp-magnets/QA_best_magmatbert_4_epochs_lr_3e-05_BS_16"
    elif model_name == "matscibert_squad":
        model_checkpoint = "nlp-magnets/QA_best_matscibert_4_epochs_lr_3e-05_BS_32"
    elif model_name == "bert_squad_Tc":
        model_checkpoint = "nlp-magnets/QA_best_Tc_finetunedQA_best_bert-base-uncased_4_epochs_lr_3e-05_BS_16_10_epochs_lr_5e-05_BS_32"
    elif model_name == "magbert_squad_Tc":
        model_checkpoint = "nlp-magnets/QA_best_Tc_finetunedQA_best_magbert_4_epochs_lr_5e-05_BS_64_10_epochs_lr_1e-05_BS_64"
    elif model_name == "magmatbert_squad_Tc":
        model_checkpoint = "nlp-magnets/QA_best_Tc_finetunedQA_best_magmatbert_4_epochs_lr_3e-05_BS_16_10_epochs_lr_3e-05_BS_16"
    elif model_name == "matscibert_squad_Tc":
        model_checkpoint = "nlp-magnets/QA_best_Tc_finetunedQA_best_matscibert_4_epochs_lr_3e-05_BS_32_10_epochs_lr_5e-05_BS_32"
        
    question_answerer = pipeline("question-answering", model = model_checkpoint, token="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")
    
    # predictions
    predicted = question_answerer(question = dataset['train']['question'], context = dataset['train']['context'])
    return predicted
```

```python
def metric_scores(predictions, GT):
    ### TP - true positives (positive predictions that are actualle positive)
    TP=0
    for true, predicted in zip(GT['train']['answers'], predictions):
        just_numeric_inpred=re.findall(r'\d+\.*\d*', predicted['answer'], re.IGNORECASE)
        if just_numeric_inpred == true['text']:
            TP += 1
        
    ### Precision - what proportion of positive identifications was actually correct?
    ## From all the positive predictions (in our case positive is numerical answer and negative is "no-answer"/or non-numerical) 
    ## we find the proportion of TPs
    pred_lst = [pred['answer'] for pred in predictions]
    pred_lst_final = []
    for pred_ in pred_lst:
        just_numeric_pred = re.findall(r'\d+\.*\d*', pred_, re.IGNORECASE)
        if len(just_numeric_pred) > 0:
            pred_lst_final.append(just_numeric_pred[0])
    precision = TP / len(pred_lst_final)

    ### Recall - what proportion of actual positives was identified correctly?
    ## From all the positive GT (in our case positive is non-zero numerical answer and negative is "0.00" or 'answer_start': [-1]) 
    ## we find the proportion of TPs
    true_lst_final = [x['text'][0] for x in GT['train']['answers'] if x['text'][0] != "0.00"]
    len(true_lst_final)
    recall = TP / len(true_lst_final)

    F1_score = 2*precision*recall / (precision + recall)

    return round(precision*100, 2), round(recall*100, 2), round(F1_score*100, 2), TP, len(pred_lst_final), len(true_lst_final)
```

```python
def metric_scores_no_answer(predictions, GT):
    ### TP - true positives (positive predictions that are actualle positive)
    TP=0
    for true,predicted in zip(GT['train']['answers'], predictions):
        just_numeric_inpred=re.findall(r'\d+\.*\d*',predicted['answer'],re.IGNORECASE)
        if just_numeric_inpred==true['text']:
            TP+=1
        
    ### Precision - what proportion of positive identifications was actually correct?
    ## From all the positive predictions (in our case positive is numerical answer and negative is "no-answer") 
    ## we find the proportion of TPs
    pred_lst = [pred['answer'] for pred in predictions]
    pred_lst_final = [x for x in pred_lst if x != "no-answer"]
    precision = TP / len(pred_lst_final)

    ### Recall - what proportion of actual positives was identified correctly?
    ## From all the positive GT (in our case positive is non-zero numerical answer and negative is "0.00" or 'answer_start': [-1]) 
    ## we find the proportion of TPs
    true_lst_final = [x['text'][0] for x in GT['train']['answers'] if x['text'][0] != "0.00"]
    len(true_lst_final)
    recall = TP / len(true_lst_final)

    F1_score = 2*precision*recall / (precision + recall)

    return round(precision*100, 2), round(recall*100, 2), round(F1_score*100, 2), TP, len(pred_lst_final), len(true_lst_final)
```
